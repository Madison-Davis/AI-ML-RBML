```python
# Imports
# Link 1: https://pypi.org/project/pyECLAT/
import pandas as pd
import numpy as np
!pip install pyECLAT
from pyECLAT import ECLAT
```

```python
# Data
# Dataset: https://www.kaggle.com/code/knightbearr/analysis-shopping-cart-data-knightbearr/data?select=sales.csv
# row[1] = Order ID
# row[2] = Product ID
dataFile = "//content//drive//MyDrive//Coding//Personal Projects//1: Artificial Intelligence//Resources//EclatApparelTransactions.csv"
data = pd.read_csv(dataFile)
data = data.values.tolist()
firstRow = True
allTransactions = []
singleTransaction = []
for row in data:
  if firstRow:
    previousRow = row[1]
    singleTransaction.append(str(row[2]))
    firstRow = False
  elif row[1] == previousRow:
    singleTransaction.append(str(row[2]))
  else:
    previousRow = row[1]
    allTransactions.append(singleTransaction)
    singleTransaction = []

allTransactions = pd.DataFrame(allTransactions)
```

```python
# ECLAT Algorithm
eclat_instance = ECLAT(data = allTransactions, verbose = True)
get_ECLAT_indexes, get_ECLAT_supports = eclat_instance.fit(min_support = 0.0008, min_combination = 2)
```
